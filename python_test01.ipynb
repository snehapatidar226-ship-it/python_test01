```python
# Question:-03

s="hello sneha"
vowels="aeiouAEIOU"
v=c=0
for ch in s:
    if ch!="":
        if ch in vowels:
            v+=1
        else:
            c+=1
print("vowels:",v)
print("consonants:", c)
```

```python
# Question -04

s="madam"
rev=""
for ch in s:
    rev=ch+rev
if s==rev:
    print("palindrome")
else:
    print("not palindrome")
```
