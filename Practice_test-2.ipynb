# String Manipulation

```python
# Question 01:-
s="My Name Is Sneha"
def print_with_dash(s):
    result = ""
    for i in s:
        result += i + "-"
    print(result)

print_with_dash(s)

    
```

```python
# Question 02:- 

def count_upper_lower(s):
    upper = 0
    lower = 0
    for i in s:
        if i.isupper():
            upper += 1
        elif i.islower():
            lower += 1
    print("Uppercase:", upper)
    print("Lowercase:", lower)

count_upper_lower(s)

```

```python
# Question 03:- 

def count_upper_lower(s):
    upper = 0
    lower = 0
    for i in s:
        if i.isupper():
            upper = upper + 1
        elif i.islower():
            lower = lower + 1
    print("Uppercase:", upper)
    print("Lowercase:", lower)

count_upper_lower(s)
```

```python
# Question 04:- 

def first_index(s, ch):
    for i in range(len(s)):
        if s[i] == ch:
            print("Index:", i)
            return
    print("Character not found")

first_index(s, "a")
```

```python
# Question 05:-

def count_char(s, ch):
    count = 0
    for i in s:
        if i == ch:
            count = count + 1
    print("Count:", count)

count_char(s, "a")
```

```python
# Question 06:- 

def remove_spaces(s):
    result = ""
    for i in s:
        if i != " ":
            result = result + i
    print(result)

remove_spaces(s)
```

```python
# Question 07:-

def replace_digits(s):
    result = ""
    for i in s:
        if i.isdigit():
            result = result + "*"
        else:
            result = result + i
    print(result)

replace_digits(s)
```

```python
# Question 08:-

def print_consonants(s):
    vowels = "aeiouAEIOU"
    result = ""
    for i in s:
        if i.isalpha() and i not in vowels:
            result = result + i
    print(result)

print_consonants(s)
```

```python
# Question 09:-
def print_vowels(s):
    vowels = "aeiouAEIOU"
    result = ""
    for i in s:
        if i in vowels:
            result = result + i
    print(result)

print_vowels(s)
```

```python
# Question 10:-
def is_palindrome(s):
    left = 0
    right = len(s) - 1
    while left < right:
        if s[left] != s[right]:
            print("Not Palindrome")
            return
        left = left + 1
        right = right - 1
    print("Palindrome")

is_palindrome("madam")
```
